```python
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
from nltk.stem import WordNetLemmatizer

lemmati = WordNetLemmatizer()

words = ["sodis"]

for i in words:
  print("i :", lemmati.lemmatize(i))
```
