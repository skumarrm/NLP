```python
import nltk
nltk.download('punkt')
```

```python
from nltk.stem import PorterStemmer

ps = PorterStemmer()

words1 = ["like", "likers", "liking", "likes" ]
words2 = ["men", "mens", "man" ]

for i in words1+words2:
  print (i, ":", ps.stem(i))
```
